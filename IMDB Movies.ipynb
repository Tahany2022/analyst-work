# preparing data to be numeric i will convert it from object  to float and make wanted replacement 

```python
movd['votes']=movd['votes'].str.replace(',','').astype(int)
movd['gross_earn']=movd['gross_earn'].str.replace('$','').str.replace('M','')
movd['gross_earn']=movd['gross_earn'].astype(float)
movd['runtime']=movd['runtime'].str.replace("min","").astype(float)
movd['title']=movd['title'].astype('category')
```

```python
sns.boxplot(movd['votes'])
```

```python
# find the types of films accordin to votes and calssified it into high,medium,low
def calssify_votes(x):
    if x>100000:
        return 'high'
    elif 10000<x<100000:
        return "medium"
    elif x<10000:
        return 'low'
movd['movie_votes']=movd['votes'].apply(calssify_votes)
movd.tail()
```

```python
# find the top 100 films according to votes
```

```python
# find the maximum two films the have the largest gross earn accrding to each director
for i in directors_list:
    print(i)
    view=directors.get_group(i).nlargest(2,'gross_earn')['title'].values
    print(view)
    
```

# visulization

```python
#here i will make a visulization to the high and low gross earn of the film
```

```python
movd['gross_earn_state']=np.where(movd['gross_earn'],'high','low')
x=movd['gross_earn_state'].value_counts()
colors=sns.color_palette('bright')
plt.pie(x.values,labels=x.index,colors=colors,autopct='%.2f%%')
plt.xticks(rotation=90)
plt.show()
```
